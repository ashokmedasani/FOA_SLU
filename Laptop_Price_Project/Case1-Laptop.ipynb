**Introduction**

In this project, we will load and analyze laptop prices data

```python
#Load the data

import pandas as pd

df = pd.read_csv("laptop_pricing_dataset_mod2-1.csv")
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
df.columns
```

```python
#Drop the unnecessary columns and take that as new df

df_clean = df.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0'])
```

```python
print(df)
```

Univariate Analysis

If Numerical -> Make Histograms

If Categorical -> Bar Graphs

```python
#Get the count of each value in the Manufacturer column

manufacturer_counts = df_clean['Manufacturer'].value_counts()

# Convert teh counts to a df to work easily

manufacturer_counts_df = pd.DataFrame(manufacturer_counts).reset_index()
manufacturer_counts_df.columns = ['Manufacturer','Count']

print(manufacturer_counts_df)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Sort the Dataframe by count in descending order

manufacturer_counts_df = manufacturer_counts_df.sort_values(by= "Count", ascending= False)

#Set the style of Seaborn
sns.set_style("whitegrid")

#Set the palette
palette = sns.color_palette('pastel')

# Create the bar graph
plt.figure(figsize= (10,6))
sns.barplot(data = manufacturer_counts_df, x= 'Manufacturer', y = 'Count', palette = palette)
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.title('Count of laptops by Manufacturer')
plt.xticks(rotation = 45)
plt.show()
```

**Histogram**

```python
sns.histplot(df_clean['Price'], kde = True, color = 'skyblue')
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show();
```

```python
#Get the count of each value in the Manufacturer column

price_binned_counts = df_clean['Price-binned'].value_counts()

# Convert teh counts to a df to work easily

price_binned_counts_df = pd.DataFrame(price_binned_counts).reset_index()
price_binned_counts_df.columns = ['Price-binned','Count']

print(price_binned_counts_df)
```

```python



# Sort the Dataframe by count in descending order

price_binned_counts_df = price_binned_counts_df.sort_values(by= "Count", ascending= False)
# print(price_binned_counts_df)
#Set the style of Seaborn
# sns.set_style("whitegrid")

#Set the palette
# palette = sns.color_palette('pastel')

# Create the bar graph
plt.figure(figsize= (6,6))
sns.barplot(data = price_binned_counts_df, x = 'Price-binned', y = 'Count', palette = palette)
plt.xlabel('Price Binned')
plt.ylabel('Count')
plt.title('Count of laptops by Price Binned')
plt.xticks(rotation = 45)
plt.show()
```
